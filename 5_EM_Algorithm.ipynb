```python
import numpy as np
```

# (a) Create Gaussian means

```python
true_means = [2, 8, 14]
variance = 1
std = np.sqrt(variance)
```

# (b) Generate data

```python
np.random.seed(0)
data = []
for m in true_means:
    samples = np.random.normal(m, std, 40)
    data.extend(samples)
data = np.array(data)

# Shuffle points
np.random.shuffle(data)
print("After shuffle data: ")
print(data)
```

# (c) EM Algorithm

```python
k = 3
n = len(data)

# random initialization of means
means = np.random.choice(data, k)
max_iter = 100
for iteration in range(max_iter):
    responsibilities = np.zeros((n, k))
    for j in range(k):
        responsibilities[:, j] = np.exp(-(data - means[j])**2 / (2*variance))

    responsibilities = responsibilities / responsibilities.sum(axis=1, keepdims=True)

    new_means = []

    for j in range(k):
        weight = responsibilities[:, j]
        new_mean = np.sum(weight * data) / np.sum(weight)
        new_means.append(new_mean)
    new_means = np.array(new_means)

    # convergence check
    if np.allclose(means, new_means, atol=1e-4):
        break

    means = new_means
```

# (d) Compare results

```python
print("True Means:", true_means)
print("Estimated Means:", means)
```
